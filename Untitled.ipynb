```python
import warnings
warnings.filterwarnings("ignore")
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import yfinance as yf
from sklearn.metrics import roc_curve, auc, roc_auc_score
from sklearn.model_selection import train_test_split, TimeSeriesSplit, GridSearchCV
from sklearn.utils.class_weight import compute_sample_weight
from sklearn.ensemble import RandomForestClassifier
from statsmodels.tsa.holtwinters import SimpleExpSmoothing
```

```python
def rsi(data, period=15):
    delta = data.diff()
    avg_gain = (delta.where(delta > 0, 0)).rolling(window=period).mean()
    avg_loss = (-delta.where(delta < 0, 0)).rolling(window=period).mean()
    rs = avg_gain / avg_loss
    rsi = 100 - (100 / (1 + rs))
    return rsi
```

```python
def stochastic_oscillator(data, period=14):
    low_min = data['Low'].rolling(window=period).min()
    high_max = data['High'].rolling(window=period).max()
    stoch = 100 * ((data['Close'] - low_min) / (high_max - low_min))
    return stoch
```

```python
def williams_r(data, period=14):
    low_min = data['Low'].rolling(window=period).min()
    high_max = data['High'].rolling(window=period).max()
    williams = -100 * ((high_max - data['Close']) / (high_max - low_min))
    return williams
```

```python
def macd(data, fast_period=12, slow_period=26):
    macd_line = data['Close'].ewm(span=fast_period).mean() - data['Close'].ewm(span=slow_period).mean()
    return macd_line
```

```python
def price_rate_of_change(data, period=14):
    roc = data['Close'].pct_change(periods=period) * 100
    return roc
```

```python
def on_balance_volume(data):
    obv = (np.sign(data['Close'].diff()) * data['Volume']).fillna(0).cumsum()
    return obv
```

```python
def apply_simple_exp_smoothing(series, alpha=0.2):
    """
    Apply Simple Exponential Smoothing to a pandas Series.
    """
    model = SimpleExpSmoothing(series, initialization_method='legacy-heuristic')
    fit_model = model.fit(smoothing_level=alpha, optimized=False)
    return fit_model.fittedvalues

def get_data(stocks, start_date, end_date, d):
    stock_data_dict = {}
    for stock in stocks:
        stock_data = yf.download(stock, start=start_date, end=end_date)
        # Apply Simple Exponential Smoothing instead of EWM
        stock_data['Adj Close'] = apply_simple_exp_smoothing(stock_data['Adj Close'])
        stock_data['High'] = apply_simple_exp_smoothing(stock_data['High'])
        stock_data['Low'] = apply_simple_exp_smoothing(stock_data['Low'])
        stock_data['Volume'] = apply_simple_exp_smoothing(stock_data['Volume'])
        data = pd.DataFrame()
        data['RSI'] = rsi(stock_data['Adj Close'])
        data['Stochastic Oscillator'] = stochastic_oscillator(stock_data)
        data['Williams %R'] = williams_r(stock_data)
        data['MACD'] = macd(stock_data)
        data['Price Rate of Change'] = price_rate_of_change(stock_data)
        data['On Balance Volume'] = on_balance_volume(stock_data)
        data[f'Target_{d}'] = np.sign(stock_data['Close'].shift(-d) - stock_data['Close'])
        data = data[data[f'Target_{d}'] != 0]
        data = data.dropna()
        stock_data_dict[stock] = data
    aapl_data = stock_data_dict['AAPL']
    samsung_data = stock_data_dict['005930.KS']
    ge_data = stock_data_dict['GE']

    return aapl_data, samsung_data, ge_data
```

```python
def best_estimator_rf(stock_data, d):
    X = stock_data.drop(columns=[f'Target_{d}'])
    y = stock_data[f'Target_{d}']
    y = y.replace(-1, 0)

    tscv = TimeSeriesSplit(n_splits=5)

    param_grid = {
        'n_estimators': [5, 25, 45, 65, 90, 100],
        'max_depth': [3, 5, 7],
        'min_samples_split': [2, 5, 10]
    }
    rf = RandomForestClassifier(random_state=42, oob_score=True)
    grid_search = GridSearchCV(rf, param_grid, cv=tscv, n_jobs=-1, verbose=1)
    grid_search.fit(X, y)

    model = grid_search.best_estimator_
    split_indices = list(tscv.split(X))
    train_indices, test_indices = split_indices[-1]
    X_train, X_test = X.iloc[train_indices], X.iloc[test_indices]
    y_train, y_test = y.iloc[train_indices], y.iloc[test_indices]

    model.fit(X_train, y_train)

    y_prob = model.predict_proba(X_test)[:, 1]
    metric = {
        'Sample Size': stock_data.shape[0],
        'Training Score': model.score(X_train, y_train)
    }

    return model, metric, X_test, y_test, y_prob
```

```python
def rf_models_fixed_estimators(stock_data, d):
    """
    Train and evaluate Random Forest using TimeSeriesSplit.
    """
    X = stock_data.drop(columns=[f'Target_{d}'])
    y = stock_data[f'Target_{d}']
    y = y.replace(-1, 0)

    print(f"\n--- Processing TimeSeriesSplit for Trading Period: {d} days ---")

    tscv = TimeSeriesSplit(n_splits=5)
    splits = list(tscv.split(X))
    train_indices, test_indices = splits[-1]
    X_train, X_test = X.iloc[train_indices], X.iloc[test_indices]
    y_train, y_test = y.iloc[train_indices], y.iloc[test_indices]

    sample_weights = compute_sample_weight('balanced', y_train)
    n_estimators_list = [5, 25, 45, 65]
    selected_model = None
    selected_metrics = None

    for n in n_estimators_list:
        model = RandomForestClassifier(n_estimators=n, random_state=42, oob_score=True)
        model.fit(X_train, y_train, sample_weight=sample_weights)

        oob_error = 1 - model.oob_score_

        print(f"Model with {n} trees -> OOB Score: {model.oob_score_:.4f}, "
              f"OOB Error: {oob_error:.4f}")

        if n == 65:  # Use the model with 65 estimators for evaluation
            selected_model = model
            selected_metrics = {
                'Sample Size': stock_data.shape[0],
                'OOB Error': oob_error,
            }

    return selected_model, selected_metrics, X_test, y_test
```

```python
def train_test_split_rf(stock_data, d, test_size=0.2):
    """
    Train and evaluate Random Forest using train_test_split.
    """
    X = stock_data.drop(columns=[f'Target_{d}'])
    y = stock_data[f'Target_{d}']
    y = y.replace(-1, 0)

    print(f"\n--- Processing Train-Test Split for Trading Period: {d} days ---")

    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=test_size, random_state=42, stratify=y
    )

    sample_weights = compute_sample_weight('balanced', y_train)

    model = RandomForestClassifier(
        random_state=42, n_estimators=100, oob_score=True
    )
    model.fit(X_train, y_train, sample_weight=sample_weights)

    oob_error = 1 - model.oob_score_

    print(f"OOB Score: {model.oob_score_:.4f}, OOB Error: {oob_error:.4f}, "
              f"OOB Error: {oob_error:.4f}")
         

    y_prob = model.predict_proba(X_test)[:, 1]

    metric = {
        'Sample Size': stock_data.shape[0],
        'OOB Error': oob_error,
    }

    return model, metric, X_test, y_test, y_prob
```

```python
def plot_roc_curves_with_datasets(models, datasets, labels, title):
    """
    Plot ROC curves for multiple models with datasets and add labels for each curve.
    """
    plt.figure(figsize=(8, 6))
    colors = ['black', 'blue', 'red']  # Define colors for each curve
    plt.plot([0, 1], [0, 1], linestyle='--', color='brown', lw=2, label="Random Baseline")

    for i, (model, (X, y)) in enumerate(zip(models, datasets)):
        y_pred_proba = model.predict_proba(X)[:, 1]
        fpr, tpr, _ = roc_curve(y, y_pred_proba)
        roc_auc = auc(fpr, tpr)
        plt.plot(fpr, tpr, color=colors[i], lw=2, label=f"ROC curve for {labels[i]} days (AUC = {roc_auc:.2f})")

    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title(title)
    plt.legend(loc="best")
    plt.grid()
    plt.show()
```

```python


stocks = ['AAPL', '005930.KS', 'GE']
start_date = '2002-01-01'
end_date = '2023-12-31'

# Initialize storage for models and datasets for each stock
models_tts = {stock: [] for stock in stocks}   # Train-test split models
models_tscv = {stock: [] for stock in stocks}  # TimeSeriesSplit models
datasets_tts = {stock: [] for stock in stocks}  # Train-test split datasets
datasets_tscv = {stock: [] for stock in stocks}  # TimeSeriesSplit datasets

# Initialize OOB error storage
oob_error_storage = {stock: {30: [], 60: [], 90: []} for stock in stocks}

labels = ['30', '60', '90']
n_estimators_list = [5, 25, 45, 65, 90, 100]

# Process each trading period
for d in [30, 60, 90]:
    print(f"\nProcessing trading period: {d} days")

    # Get data for the trading period `d`
    aapl_data, samsung_data, ge_data = get_data(stocks, start_date, end_date, d)

    # Create a dictionary for the datasets
    stock_data_dict = {
        'AAPL': aapl_data,
        '005930.KS': samsung_data,
        'GE': ge_data
    }

    for stock, data in stock_data_dict.items():
        if data.empty:
            print(f"Skipping {stock} for {d} days - no data available.")
            continue

        print(f"\nProcessing {stock} for {d}-day trading period...")

        # Train-Test Split
        model_tts, metric_tts, X_test_tts, y_test_tts, y_prob_tts = train_test_split_rf(data, d)
        models_tts[stock].append(model_tts)
        datasets_tts[stock].append((X_test_tts, y_test_tts))

        # TimeSeriesSplit
        selected_model_tscv, selected_metric_tscv, X_test_tscv, y_test_tscv = rf_models_fixed_estimators(data, d)
        models_tscv[stock].append(selected_model_tscv)  # Append the selected model directly
        datasets_tscv[stock].append((X_test_tscv, y_test_tscv))

        # Calculate OOB Errors for different numbers of estimators
        target_col = f'Target_{d}'
        X = data.drop(columns=[target_col])
        y = data[target_col].replace(-1, 0)
        sample_weights = compute_sample_weight('balanced', y)

        for n_estimators in n_estimators_list:
            try:
                model = RandomForestClassifier(n_estimators=n_estimators, random_state=42, oob_score=True)
                model.fit(X, y, sample_weight=sample_weights)

                # Compute and store OOB error
                oob_error = 1 - model.oob_score_
                oob_error_storage[stock][d].append(oob_error)

                print(f"Stock {stock} - Trading Period {d} days - n_estimators {n_estimators}: OOB Error Rate = {oob_error:.4f}")
            except ValueError as e:
                print(f"Error training model for {stock} with {n_estimators} estimators: {e}")
                break

# 1. Block for ROC Curve Plotting
def plot_roc_curves_with_datasets(models, datasets, labels, title):
    """
    Plot ROC curves for multiple models with datasets and add labels for each curve.
    """
    plt.figure(figsize=(8, 6))
    plt.plot([0, 1], [0, 1], linestyle='--', color='brown', lw=2, label="Random Baseline")

    for i, (model, (X, y)) in enumerate(zip(models, datasets)):
        y_pred_proba = model.predict_proba(X)[:, 1]
        fpr, tpr, _ = roc_curve(y, y_pred_proba)
        roc_auc = auc(fpr, tpr)
        plt.plot(fpr, tpr, lw=2, label=f"ROC curve for {labels[i]} days (AUC = {roc_auc:.2f})")

    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title(title)
    plt.legend(loc="best")
    plt.grid()
    plt.show()

# Call ROC Plotting Block
for stock in stocks:
    plot_roc_curves_with_datasets(models_tts[stock], datasets_tts[stock], labels, f'{stock} stock price (Train-Test Split)')
    plot_roc_curves_with_datasets(models_tscv[stock], datasets_tscv[stock], labels, f'{stock} stock price (TimeSeriesSplit)')


```

```python
# 2. Block for OOB Error Plotting
def plot_oob_errors(oob_error_storage, n_estimators_list, stocks):
    """
    Plot OOB error rates vs. number of estimators for each stock and trading period.
    """
    for stock in stocks:
        plt.figure(figsize=(10, 7))
        for d, errors in oob_error_storage[stock].items():
            # Ensure lengths match for plotting
            n_estimators_trimmed = n_estimators_list[:len(errors)]
            plt.plot(
                n_estimators_trimmed,
                errors,
                marker='o',
                linestyle='-',
                label=f"{d}-day trading period"
            )
        
        plt.xlabel("Number of Estimators")
        plt.ylabel("OOB Error Rate")
        plt.title(f"OOB Error Rate vs. Number of Estimators for {stock}")
        plt.legend(loc="best")
        plt.grid()
        plt.show()

# Call OOB Plotting Block
plot_oob_errors(oob_error_storage, n_estimators_list, stocks)
```
